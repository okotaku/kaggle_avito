```python
import pandas as pd
import glob

path = glob.glob("/Volumes/USB/kaggle/avito/val/f*.csv")
path = glob.glob("/Volumes/USB/kaggle/avito/val/*/bagging.csv")
for i, p in enumerate(path):
    df = pd.read_csv(p)
    df.rename(columns={"deal_probability": p.replace("val/", "").replace(".csv", "")}, inplace=True)
    if i == 0:
        df_ = df
    else:
        df_ = df_.merge(df, on="item_id")
        
df_.drop("item_id", axis=1, inplace=True)
df_corr = df_.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(20, 20))
sns.heatmap(df_corr, annot=True)
plt.savefig('corr.png')
```

```python
df_corr.to_csv("corr.csv")
```
